```python
import os
import openai
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_fighter_style(fighter_name):
    key = os.environ["Open_API_Key"]
    openai.api_key = key

    prompt = f"Provide martial arts fighting style of {fighter_name}. Restrict answers to ONE of: Boxing, Brazilian Jiu-Jitsu, Sambo, Wrestling, Judo, Karate, Kung-Fu, Taekwondo"

    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        temperature=0.0,
        max_tokens=250,
        n=1,
        frequency_penalty=0,
        presence_penalty=0,
        stop= None,
    )

    return response.choices[0].text.strip()
```

```python
fighter_name = "Connor McGregor"
training_history = get_fighter_style(fighter_name)
print(training_history)
```

```python
all_stats = pd.read_csv('data/final/ALL_Stats_and_Bios.csv')
all_stats.head()
```

```python
# get list of fighters
fighter_list = all_stats['Fighter_Name'].unique()
```

```python
# test
fighter1 = fighter_list[0]
fighter_style = get_fighter_style(fighter1)
# make df, add fighters name as one col, then add training history as another col
fighter_style_df = pd.DataFrame(columns=['Fighter_Name', 'Fighter_Style'])
fighter_style_df= fighter_style_df.append({'Fighter_Name': fighter1, 'Fighter_Style': fighter_style}, ignore_index=True)
# add fighter name to row
fighter_style_df['Fighter_Name'] = fighter1
# move fighter name to first column
fighter_style_df
```

```python
# Check to see which fighters have already been downloaded
fighter_styles_downloaded = os.listdir('data/final/fighter_styles')
fighter_styles_downloaded = [x.replace('.csv', '') for x in fighter_styles_downloaded]

# find fighters that have not been downloaded yet
fighters_to_download = [x for x in fighter_list if x not in fighter_styles_downloaded]

all_fighter_styles = pd.DataFrame(columns=['Fighter_Name', 'Training_History'])

i = 1

for fighter in fighters_to_download:
    style = get_fighter_style(fighter)
    # make df, add fighters name as one col, then add training history as another col
    fighter_style_df = pd.DataFrame(columns=['Fighter_Name', 'Fighter_Style'])
    fighter_style_df= fighter_style_df.append({'Fighter_Name': fighter, 'Fighter_Style': style}, ignore_index=True)
    # add fighter name to row
    fighter_style_df['Fighter_Name'] = fighter

    # save fighter training history to csv
    fighter_style_df.to_csv(f'data/final/fighter_styles/{fighter}.csv', index=False)
    print(f'{i} of {len(fighters_to_download)} complete')
    i += 1

    all_fighter_styles = all_fighter_styles.append(fighter_style_df, ignore_index=True)

all_fighter_styles
```

```python
# drop training history column
all_fighter_styles.drop(columns='Training_History', inplace=True)
```
